```python
# Load the NetCDF file
loaded_ds = xr.load_dataset('./Data/BT_02degree_SGP_SEUS.nc')
BT_SGP_01 = loaded_ds['BT_SGP_01'].values # save Brightness Temperatures with in 0.2 and 0.2 domain 
BT_SEUS_01 = loaded_ds['BT_SEUS_01'].values
```

```python
bins=np.arange(11)*5-10
bin_center=np.arange(10)*5-10+2.5
```

```python
data1_01=BT_SGP_01[ (BT_SGP_01>200) & (BT_SGP_01<999) ]
data2_01=BT_SEUS_01[ (BT_SEUS_01>200) & (BT_SEUS_01<999) ]

counts1_01, _ = np.histogram(data1_01-273.15, bins=bins)
percentages1_01 = counts1_01 / counts1_01.sum() * 100

counts2_01, _ = np.histogram(data2_01-273.15, bins=bins)
percentages2_01 = counts2_01 / counts2_01.sum() * 100
```

```python

mpl.rcParams['font.size'] = 24
mpl.rcParams['font.weight'] = 'regular'
mpl.rcParams['font.family'] = 'DejaVu Sans'
fig = plt.figure(figsize=[18,9], dpi=40)

gs = gridspec.GridSpec(1, 20)
gs.update(wspace=0.5, hspace=0.3, top=0.91)

ax2 = plt.subplot(gs[0,0:20])


######
ax2.plot(bin_center, percentages1_01,'-o', label='SGP',color='blue',linewidth=5,markersize=20)
ax2.plot(bin_center, percentages2_01,'-o', label='Alabama',color='red',linewidth=5,markersize=20)
ax2.set_xlabel('Brightness Temperature ($^\circ$C)')
ax2.set_ylabel('PDF (%)')
ax2.grid()
ax2.legend()
ax2.set_xticks([0, 5, 10, 15, 20,25,30])
ax2.set_xlim(0,35)


#fig.savefig(f'./SGP_Alabama_BrightnessTemperature.png', dpi=300, bbox_inches='tight')
```
